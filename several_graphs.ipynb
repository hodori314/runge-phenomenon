```python
import numpy as np
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial
```

```python
# degree 3
x_3 = []; y_3=[]
n= 3
for i in range(n+1):
    temp = ((2*i)/n)-1
    x_3.append(temp)
    y_3.append(1/(1+25*temp*temp))
print(x_3)
print(y_3)

poly_3 = lagrange(x_3, y_3)

# degree 7
x_7 = []; y_7=[]
n = 7
for i in range(n+1):
    temp = ((2*i)/n)-1
    x_7.append(temp)
    y_7.append(1/(1+25*temp*temp))
print(x_7)
print(y_7)
poly_7 = lagrange(x_7, y_7)

# degree 15
x_15 = []; y_15=[]
n= 15
for i in range(n+1):
    temp = ((2*i)/n)-1
    x_15.append(temp)
    y_15.append(1/(1+25*temp*temp))
print(x_15)
print(y_15)

poly_15 = lagrange(x_15, y_15)
```

```python
# graph

x_new = np.arange(-1.0, 1.01, 0.01)
plt.plot(x_new, Polynomial(poly_3.coef[::-1])(x_new), 'b', label='3th Lagrange Interpolation Polynomial')
plt.plot(x_new, Polynomial(poly_7.coef[::-1])(x_new), 'g', label='7th Lagrange Interpolation Polynomial')
plt.plot(x_new, Polynomial(poly_15.coef[::-1])(x_new), 'r', label='15th Lagrange Interpolation Polynomial')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.figure(figsize=(6,8))
plt.show()
```

```python
# scatter graph
x_new = np.arange(-1.0, 1.01, 0.01)
plt.scatter(x_15, y_15, label='15th data')
plt.scatter(x_7, y_7, label='7th data')
plt.scatter(x_3, y_3, label='3th data')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.figure(figsize=(6,8))
plt.show()
```

```python
# graph with n=15 Lagrange Polynomial
y_=[];
for i in x_new:
    temp = 1/(1+25*i*i)
    y_.append(temp)
plt.plot(x_new,y_)
plt.scatter(x_15, y_15, label='data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# chebyshev node
semi_y=[]
for i in x_new:
    if i>=1 or i<=-1:
        temp = 0
    else: 
        temp = (1-i*i)**0.5
    semi_y.append(temp)

chev_x =[]; chev_y =[]; chev_uy=[]    
n=15
for i in range(1, n+2):
    temp = np.cos(np.pi*(2*i-1)/(2*n))
    chev_x.append(temp)
    temp_ = (1-temp*temp)**0.5
    chev_y.append(temp_)
    chev_uy.append(0)

for i in range(1,n+2):
    plt.axvline(x=chev_x[i-1], ymax=chev_y[i-1], color='gray')
plt.plot(x_new, semi_y)
plt.scatter(chev_x, chev_y, color='r')
plt.scatter(chev_x, chev_uy, color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(y=0.0)
plt.show()
```
